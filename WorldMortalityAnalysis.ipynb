# **Mortality Rate due to COVID 19 Analysis:**

Which country has the highest death rate due to COVID-19 in 2020?

```python
mean4=df1['Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average'].mean()
df1['Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average']=df1['Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average'].fillna(mean4)
```

```python
fig = plt.figure(figsize=(15,6))
sns.barplot(x=df1.groupby(['Country name'])['Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average'].sum().index,y=df1.groupby(['Country name'])['Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average'].sum().values)
plt.tick_params(axis='x',rotation = 90,labelsize=7)
sns.despine()
plt.suptitle('Country wise COVID -19 deaths in 2020',weight='bold',fontname='monospace',fontsize=20)
```

From this plot we can see that Peru with 200,000 COVID deaths in a population of less than 33 million,is the most affected Country due to COVID-19 followed by America with more excess deaths.

```python
moredeath  =df1.groupby(['Country name'])['Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average'].sum().reset_index().sort_values('Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average',ascending=False).head(10)
```

```python
fig = plt.figure(figsize=(15,4))
sns.barplot(x=moredeath['Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average'],y=moredeath['Country name'])
plt.tick_params(axis='x')
plt.xlabel('')
sns.despine()
plt.suptitle('Top 10 Countries Most Affected by COVID-19',weight='bold',fontname='monospace',fontsize=20)
```

By comparing the deathrate in 2020 with Relative to 2017-2019 Average deathrate, Peru has the highest excess deaths in 2020 alone. It's deathrate has increased tremendously.

```python
data = df1.loc[:,['Country name','COVID-19 deaths per 100,000 population in 2020']]

datasort = data.sort_values(ascending=False, by='COVID-19 deaths per 100,000 population in 2020').head(10)

fig = px.bar(datasort,
             x='Country name',
             y='COVID-19 deaths per 100,000 population in 2020'
             )

fig.update_layout(title_text='Top10 countries with most deaths per 100,000 population')
fig.show()
```

Unlike the previous graph, here we compared the deathrate PER 1 MILLION Population, and therefore with Belgium's low population, the per million deathrate is higher than the rest of the countries.
As India is one of the most populated countries in the world, the per million death rate is less compared to most countries and ranked 80th.

```python
fig = px.histogram(df1,
                   x='Index of exposure to COVID-19  infections in other countries as of March 31',
                   color='Country name',
                   nbins=5
                  )

fig.update_layout(bargap=0.2, title_text='Count of index of exposure of covid19 infections')
fig.show()
```

Again, because of Low population and high deathrate per million population, Belgium also has the highest exposure to COVID 19 with the Index of 6.343.

**How are some countries having less Mortality Rate?**

Covid-19 Deaths in Island countries vs Covid-19 Deaths in Non-island countries:

```python
f = plt.figure(figsize=(8, 8))
sns.boxplot(x = 'Island', y = 'COVID-19 deaths per 100,000 population in 2020',data=df1)
```

We notice that there are More number of deaths due to Covid-19 in non-island countries than island countries.

```python
fig = px.scatter(df1,
                 x='COVID-19 deaths per 100,000 population in 2020',
                 y='Gini coefficient of income',
                 color='Country name',
                 trendline='ols',
                 trendline_scope='overall',
                 trendline_color_override='black'
                )
fig.show()
```

We can see that as the gini coefficient decreases, the covid deaths per million falls. A lower gini coefficient corresponds to less wealth disparity. We can see that European countries have more fair wealth distribution than African countries.
But on Contrary, African countries has less deathrate compared to European, Hence, we cannot conclude that Deathrate and Gini Coefficient of income are relatable.

```python
fig = px.scatter(df1,
                 x='Population 2020',
                 y='COVID-19 deaths per 100,000 population in 2020',
                 color='Country name',
                 trendline='ols',
                 trendline_scope='overall',
                 trendline_color_override='black'
                )
fig.show()
```

Belgium is having *Low* population but has high death rate, India and china are having *High* population but low death rate compared to remaing countries and most of the countries are clustered at a corner.

```python
fig = px.scatter(df1,
                 x='Index of institutional trust',
                 y='COVID-19 deaths per 100,000 population in 2020',
                 color='Country name',
                 trendline='ols',
                 trendline_scope='overall',
                 trendline_color_override='black'
                )
fig.show()
```

From this plot we see that, for lesser index of institutional trust, the death rate is both low and high.

But As the Index of institutional trust increases, the deathrate is less based on trends.
